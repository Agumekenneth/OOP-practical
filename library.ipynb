```python
class Book:
    def __init__(self,title, author, copies_available):
        self.title = title
        self.author = author
        self.copies_available = copies_available
        self.copies = {}
    def new_count(self, new_count):
        self.copies_available += new_count
    def borrow_books(self):
        if self.copies_available >0:
            self.copies_available-= 1
            print(f"we are left with {self.copies_available} books")
        else:
            print(f"The book with {self.title}dosen't exist or is currently unavailable")
    def return_books(self):
        self.copies_available += 1
        print(f"The stock of books has been updated {self.copies_available} copies")
Book1 = Book("Alex Rider","Harper lee",4)
Book1.borrow_books()
Book1.return_books()
Book1.new_count(4)
print(Book1.copies_available)


```
