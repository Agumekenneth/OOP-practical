```python
class Course():
    def __init__(self, course_name,course_code):
        self.course_name= course_name
        self.course_code= course_code
    def __str__(self):
        return (f"{self.course_code} :{self.course_name}")

class Student():
    def __init__(self, name,age, student_id ):
        self.name= name
        self.age= age
        self.student_id= student_id
        self.enrolled_courses= []
    #Enrollment of the courses
    def enroll(self, course_name):
        for course in self.enrolled_courses:
            if course.course_name.lower() == course_name.lower():
              print(f"{self.name}is in enrolled in {course_name}")
              return 
        # Create a new course (we don't have code, so use "N/A")
        new_course = Course(course_name, "N/A")
        self.enrolled_courses.append(new_course)
        print(f"{self.name} successfully enrolled in {course_name}")

    def drop(self,course_name):
        for course in self.enrolled_courses:
            if course.course_name.lower()==course_name.lower():
                self.enrolled_courses.remove(course)
                print(f"{self.name} has dropped course{course_name}")
                return
         # If we reach here, course was not found → edge case handled
        print(f"{self.name} is not enrolled in {course_name} - cannot drop")
    #Method to display info

    
    def display_student_info(self):
        print("="*40)
        print(f"Student Name : {self.name}")
        print(f"Student ID   : {self.student_id}")
        print(f"Age          : {self.age}")
        print("Enrolled Courses:")
        if self.enrolled_courses:
            for course in self.enrolled_courses:
                print(f"  → {course}")
        else:
            print("  None")
        print("="*40)


# ========================
# MAIN PROGRAM - DEMONSTRATION
# ========================

# Create two student objects
student1 = Student("Alice Nakato", 21, "S21B13/001")
student2 = Student("John Musa", 23, "S21B13/045")

# Enroll each student in at least two courses
print("ENROLLING STUDENTS...")
student1.enroll("Python Programming")
student1.enroll("Data Structures")
student1.enroll("Web Development")

student2.enroll("Database Systems")
student2.enroll("Python Programming")
student2.enroll("Computer Networks")

# Display initial info
print("\nINITIAL STUDENT INFORMATION")
student1.display_student_info()
student2.display_student_info()

# Drop one course for one student
print("\nDROPPING COURSES...")
student1.drop("Data Structures")        # Should succeed
student1.drop("Mathematics")            # Edge case: not enrolled → handled

# Drop a course that doesn't exist (edge case)
student2.drop("Artificial Intelligence")

# Final display after dropping
print("\nFINAL STUDENT INFORMATION")
student1.display_student_info()
student2.display_student_info()
```

```python
# =========================================
# QUESTION THREE - RESTAURANT ORDER SYSTEM
# =========================================

# a) Base class
class MenuItem:
    def __init__(self, name, price, available=True):
        self.name = name
        self.price = price
        self.available = available
    
    # This method will be overridden
    def order(self):
        if self.available:
            print(f"Ordering: {self.name} - Price: UGX {self.price:,}")
        else:
            print(f"Sorry, {self.name} is NOT AVAILABLE right now.")

# a) ii. Derived class Drink
class Drink(MenuItem):
    def __init__(self, name, price, size, available=True):
        super().__init__(name, price, available)
        self.size = size  # e.g., Small, Medium, Large
    
    # Override the order() method
    def order(self):
        if self.available:
            print(f"Ordering Drink: {self.name} ({self.size}) - Price: UGX {self.price:,}")
        else:
            print(f"Sorry, {self.name} ({self.size}) is currently not available.")

# a) iii. Derived class Food
class Food(MenuItem):
    def __init__(self, name, price, is_vegetarian, available=True):
        super().__init__(name, price, available)
        self.is_vegetarian = is_vegetarian
    
    # Override the order() method
    def order(self):
        veg_status = "Vegetarian" if self.is_vegetarian else "Non-Vegetarian"
        if self.available:
            print(f"Ordering Food: {self.name} [{veg_status}] - Price: UGX {self.price:,}")
        else:
            print(f"Sorry, {self.name} is currently not available.")

# c) Order class to manage items in one order
class Order:
    def __init__(self):
        self.items = []        # List to hold MenuItem objects (Drinks or Food)
        self.total_price = 0
    
    def add_item(self, menu_item):
        if menu_item.available:
            self.items.append(menu_item)
            self.total_price += menu_item.price
            menu_item.order()  # Calls the overridden version
        else:
            print(f"Cannot add {menu_item.name} - It is not available!")
    
    def remove_item(self, item_name):
        for item in self.items:
            if item.name.lower() == item_name.lower():
                self.items.remove(item)
                self.total_price -= item.price
                print(f"Removed {item.name} from order.")
                return
        print(f"Item '{item_name}' not found in your order.")
    
    # d) Display final order and total
    def display_order(self):
        print("\n" + "="*55)
        print("                YOUR FINAL ORDER")
        print("="*55)
        if self.items:
            for item in self.items:
                # Use polymorphism - calls correct order() based on type
                if isinstance(item, Drink):
                    veg = ""
                elif isinstance(item, Food):
                    veg = " [Veg]" if item.is_vegetarian else " [Non-Veg]"
                else:
                    veg = ""
                print(f"→ {item.name}{veg} - UGX {item.price:,}")
            print("-"*55)
            print(f"TOTAL AMOUNT TO PAY: UGX {self.total_price:,}")
        else:
            print("Your order is empty.")
        print("="*55)


# =========================================
# MAIN PROGRAM - DEMONSTRATION
# =========================================

# Create menu items
print("WELCOME TO UCU CAFETERIA SYSTEM\n")
print("Adding items to the menu...\n")

soda = Drink("Coca Cola", 3000, "Large")
juice = Drink("Fresh Mango Juice", 5000, "Medium")
water = Drink("Mineral Water", 2000, "Small", available=False)  # Not available

chicken = Food("Grilled Chicken", 15000, False)
pilau = Food("Beef Pilau", 12000, False)
chips = Food("Chips & Salad", 8000, True)

# Create a customer order
customer_order = Order()

print("Customer is placing an order...\n")
customer_order.add_item(soda)
customer_order.add_item(juice)
customer_order.add_item(chicken)
customer_order.add_item(pilau)
customer_order.add_item(water)      # This should say not available
customer_order.add_item(chips)

print("\nCustomer wants to remove one item...")
customer_order.remove_item("Fresh Mango Juice")
customer_order.remove_item("Pizza")  # Not in order

# Final order display
customer_order.display_order()
```

```python
# =========================================
# QUESTION FOUR - SMART HOME AUTOMATION SYSTEM
# =========================================

# a) i. Base class Device
class Device:
    def __init__(self, device_name, location):
        self.device_name = device_name
        self.location = location
        self.status = "off"   # Default status is off
    
    def __str__(self):
        return f"{self.device_name} in {self.location} → Status: {self.status.upper()}"

# a) ii. SmartHomeController class
class SmartHomeController:
    def __init__(self):
        self.devices = []              # List to store all devices
        self.total_devices = 0         # Tracks total number of devices added
    
    # b) i. Add a new device
    def add_device(self, device_name, location):
        new_device = Device(device_name, location)
        self.devices.append(new_device)
        self.total_devices += 1
        print(f"Added: {new_device.device_name} in {new_device.location}")
    
    # b) ii. Turn ON a device
    def turn_on_device(self, device_name):
        for device in self.devices:
            if device.device_name.lower() == device_name.lower():
                device.status = "on"
                print(f"SUCCESS: {device.device_name} in {device.location} is now ON")
                return
        # Edge case: device not found
        print(f"ERROR: Device '{device_name}' not found! Cannot turn on.")
    
    # b) iii. Turn OFF a device
    def turn_off_device(self, device_name):
        for device in self.devices:
            if device.device_name.lower() == device_name.lower():
                device.status = "off"
                print(f"SUCCESS: {device.device_name} in {device.location} is now OFF")
                return
        # Edge case: device not found
        print(f"ERROR: Device '{device_name}' not found! Cannot turn off.")
    
    # e) Display all devices and their status
    def display_all_devices(self):
        print("\n" + "="*60)
        print("           SMART HOME DEVICE STATUS")
        print("="*60)
        print(f"Total Devices in System: {self.total_devices}")
        print("-"*60)
        if self.devices:
            for device in self.devices:
                status_icon = "ON" if device.status == "on" else "OFF"
                print(f"→ {device.device_name:20} | Location: {device.location:15} | {status_icon}")
        else:
            print("No devices added yet.")
        print("="*60 + "\n")


# =========================================
# MAIN PROGRAM - FULL DEMONSTRATION
# =========================================

print("SMART HOME AUTOMATION SYSTEM - UCU HOSTELS")
print("Initializing home controller...\n")

# Create the smart home controller
home = SmartHomeController()

# c) Add three devices to the system
print("ADDING THREE DEVICES...")
home.add_device("Living Room Light", "Living Room")
home.add_device("Bedroom Fan", "Bedroom")
home.add_device("Kitchen Bulb", "Kitchen")

# Display initial status
home.display_all_devices()

# Simulate turning devices on and off
print("TURNING DEVICES ON...")
home.turn_on_device("Living Room Light")
home.turn_on_device("Bedroom Fan")
home.turn_on_device("TV")                    # Edge case: not found
home.turn_on_device("Kitchen Bulb")

# Show updated status
home.display_all_devices()

# Now turn some off
print("TURNING SOME DEVICES OFF...")
home.turn_off_device("Bedroom Fan")
home.turn_off_device("Fridge")               # Edge case: not found

# Final status
print("FINAL SMART HOME STATUS")
home.display_all_devices()
```

```python
class BankLoan:
    def __init__(self, customer_name, loan_amount, interest_rate, years=5):
        if loan_amount <= 0 or interest_rate < 0 or interest_rate > 50:
            raise ValueError("Invalid loan amount or interest rate")
        self.customer_name = customer_name
        self.loan_amount = loan_amount
        self.interest_rate = interest_rate / 100
        self.years = max(1, years)

    def calculate_monthly_payment(self):
        monthly_rate = self.interest_rate / 12
        months = self.years * 12
        if monthly_rate == 0:
            return self.loan_amount / months
        payment = self.loan_amount * (monthly_rate * (1 + monthly_rate)**months) / ((1 + monthly_rate)**months - 1)
        return round(payment, 2)

# Demo
loan = BankLoan("Alice", 500000, 7.5, 5)
print(f"Monthly Payment: KES {loan.calculate_monthly_payment()}")
```

```python
import re
from datetime import datetime

class Parking:
    def __init__(self, plate_number, owner):
        if not re.match(r"^[A-Z]{3}\d{3}[A-Z]$", plate_number):
            raise ValueError("Plate must be like KDD123A")
        self.plate_number = plate_number.upper()
        self.owner = owner
        self.entry_time = datetime.now()

    def compute_duration(self):
        duration = datetime.now() - self.entry_time
        hours = duration.total_seconds() // 3600
        mins = (duration.total_seconds() % 3600) // 60
        return f"{int(hours)}h {int(mins)}m"

# Demo
car = Parking("KDD567T", "Prof. Omondi")
print(f"Car {car.plate_number} parked by {car.owner}")
print(f"Time inside: {car.compute_duration()}")
```

```python
class SMS:
    def __init__(self, sender, receiver, message):
        if len(message) > 160:
            raise ValueError("Message exceeds 160 characters")
        self.sender = sender
        self.receiver = receiver
        self.message = message

    def send(self):
        print("SMS SENT!")
        print(f"From: {self.sender}")
        print(f"To: {self.receiver}")
        print(f"Message: {self.message}")
        print(f"({len(self.message)}/160 chars)")

# Demo
sms = SMS("0722123456", "0788112233", "Hello! How are you today?")
sms.send()
```

```python
class Song:
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist = artist
        self.duration = duration

    def __str__(self):
        return f"{self.title} - {self.artist} ({self.duration})"

class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []

    def add_song(self, song):
        if any(s.title == song.title and s.artist == song.artist for s in self.songs):
            print("Song already in playlist!")
        else:
            self.songs.append(song)
            print(f"Added: {song}")

    def show_playlist(self):
        print(f"\n=== {self.name} ===")
        for i, song in enumerate(self.songs, 1):
            print(f"{i}. {song}")
        print(f"Total songs: {len(self.songs)}\n")

# Demo
s1 = Song("Perfect", "Ed Sheeran", "4:23")
s2 = Song("Shape of You", "Ed Sheeran", "3:53")
s3 = Song("Hallelujah", "Leonard Cohen", "4:38")

playlist = Playlist("My Favorites")
playlist.add_song(s1)
playlist.add_song(s2)
playlist.add_song(s1)  # Duplicate prevented
playlist.add_song(s3)
playlist.show_playlist()
```

```python
class ResearchProject:
    def __init__(self, title, supervisor):
        self.title = title
        self.supervisor = supervisor
        self.progress = 0

    def update_progress(self, increment):
        if increment < 0:
            print("Increment cannot be negative")
            return
        self.progress = min(100, self.progress + increment)
        print(f"Progress updated: {self.progress}%")

    def status(self):
        print(f"Project: {self.title}")
        print(f"Supervisor: {self.supervisor}")
        print(f"Progress: {self.progress}%")

# Demo
proj = ResearchProject("AI in Healthcare", "Dr. Achieng")
proj.update_progress(40)
proj.update_progress(70)  # Will cap at 100
proj.status()
```

Question 20 Taxi Fare calculate

```python
class Taxi:
    def __init__(self, passenger_name, distance_km):
        if distance_km <= 0:
            raise ValueError("Distance must be positive")
        self.passenger_name = passenger_name
        self.distance_km = distance_km
        self.cost_per_km = 120

    def compute_fare(self):
        total = self.distance_km * self.cost_per_km
        print("TAXI FARE RECEIPT")
        print("-" * 25)
        print(f"Passenger: {self.passenger_name}")
        print(f"Distance: {self.distance_km} km")
        print(f"Rate: KES {self.cost_per_km}/km")
        print(f"TOTAL FARE: KES {total:,}")
        print("-" * 25)

# Demo
taxi = Taxi("John Maina", 28)
taxi.compute_fare()
```
